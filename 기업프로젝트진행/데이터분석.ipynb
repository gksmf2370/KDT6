```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
```

```python
train_data = df[df['scale_pv'] != 0]
test_data = df[df['scale_pv'] == 0]

X_train = train_data[['k_rpm_pv', 'c_temp_pv', 'n_temp_pv', 's_temp_pv']]
y_train = train_data['scale_pv']
X_test = test_data[['k_rpm_pv', 'c_temp_pv', 'n_temp_pv', 's_temp_pv']]

model = RandomForestRegressor()
model.fit(X_train, y_train)
predicted_weights = model.predict(X_test)


df.loc[df['scale_pv'] == 0, 'scale_pv'] = predicted_weights
```
