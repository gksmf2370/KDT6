```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# csv 파일 불러오는 함수
def read_csv_file(file_path):
    data = pd.read_csv(file_path)
    return data
```

```python
seoul_data = read_csv_file('seoul.csv.csv')
seoul_data.head()
```

```python
seoul_data['승하차객수']
```

```python
count = seoul_data.shape[0] # 19102
df_total_count_of_people = seoul_data['승하차객수']
```

```python
stations = set()
print(stations)
```

```python
for i in range(0, count):
    stations.add(seoul_data.iloc[i]['역명'])
    
```

```python
dict_count_of_people_on_station = dict()
for station in stations:
    dict_count_of_people_on_station[station] = 0
print(dict_count_of_people_on_station)
```

```python
for i in range(0, count):
    current = seoul_data.iloc[i]
    station = current['역명']
    dict_count_of_people_on_station[station] = dict_count_of_people_on_station[station] + current['승하차객수']
print(dict_count_of_people_on_station)
```

```python
# 1. 각 지하철 역별 평균 이용객수
for station in stations:
    dict_count_of_people_on_station[station] = dict_count_of_people_on_station[station]/2
print(dict_count_of_people_on_station)
```

```python
def std_dev_by_station(seoul_data, target_station):
    station_data_list = (seoul_data[seoul_data['역명'] == target_station])

    f_item = target_station
    f_data = station_data_list

    np_data = f_data.select_dtypes(include=np.number)
    np_data_values = np_data.values
    np_data_values

    np_data_count = len(np_data_values)
    count_list = []
    for i in range(0, np_data_count):
        count_list.append(np_data_values[i][3])
    
    std_dev = np.std(count_list)

    return (f_item, std_dev)
```

```python
# 2. 각 지하철 역별 이용객수의 표준 편차
for station in stations:
    f_item, std_dev = std_dev_by_station(seoul_data, station)
    print(f_item, std_dev)
```

```python
station_name_list = []
people_count_of_station_list = []
dict_list = dict()
for station in stations:
    station_name_list.append(station)
    station_people_sum = (seoul_data[seoul_data['역명'] == station]['승하차객수']).sum()
    people_count_of_station_list.append(station_people_sum)
    dict_list[station] = station_people_sum

print(dict_list)
```

```python
#7 이용객 수의 분포를 히스토그램으로 시각화
_df = pd.DataFrame(people_count_of_station_list, index=station_name_list)
_df.hist()
```

```python
#3.가장 많은 이용객이 있는 역의 이용객 수
max_people_count = max(dict_list.values())
print(f"가장 많은 이용객수 : {max_people_count}")
```

```python
#4. 가장 적은 이용객이 있는 역의 이용객수

min_people_count = min(dict_list.values())
print(f"가장 적은 이용객수:  {min_people_count}")
```

```python
#5. 이용객이 가장 많은 역의 이름
max_station = max(dict_list, key=dict_list.get)
print(f'이용객 수가 가장 많은 역의 이름 : {max_station}')
```

```python
#6. 이용객이 가장 많은 역의 이름
min_station = min(dict_list, key=dict_list.get)
print(f'이용객 수가 가장 적은 역의 이름 : {min_station}')
```
