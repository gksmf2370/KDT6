```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
file_path = r'C:\Users\KDP15\Desktop\KDT6\numpy통계\DAY01\boston.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
price = df['MEDV'].values
```

```python
# 1. 주택 가격의 평균
mean_price=np.mean(price)
print(mean_price)
```

```python
# 2. 주택 가격의 중앙값
median_price = np.median(price)
print(median_price)
```

```python
# 3. 주택 가격의 표준 편차
std_price = np.std(price)
print(std_price)
```

```python
# 4. 주택 가격의 최대값
max_price=np.max(price)
print(max_price)
```

```python
# 5. 주택 가격의 최소값
min_price = np.min(price)
print(min_price)
```

```python
# 6. 주택 가격의 최빈값
mode_price = df['MEDV'].mode()[0]
print(mode_price)
```

```python
# 7. 주택 가격의 분포를 시각화하세요.
# 주택 가격 데이터 (MEDV 컬럼)
prices = df['MEDV']

# 히스토그램 그리기
plt.figure(figsize=(10, 6))
plt.hist(prices, bins=30)
plt.title('주택 가격 히스토그램',fontsize=20)
plt.xlabel('가격')
plt.ylabel('분포')
plt.show()
```

```python
taxs = df['TAX']
plt.figure(figsize=(10, 6))
plt.scatter(taxs, prices, alpha=0.5, c=taxs, cmap='spring')
plt.title('세금 주택가격 산점도', fontsize=20)
plt.xlabel('세금')
plt.ylabel('주택가격')
plt.show()
```
