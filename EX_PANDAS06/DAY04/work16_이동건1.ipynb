## 타이타닉 생존자 분석

```python
# [1] 모듈 로딩
import pandas as pd
import numpy as np
```

```python
# [2] 데이터 준비
file_name='titanic.csv'
```

```python
# [3] csv ===> DataFrame으로 변환
titanDF=pd.read_csv(file_name)
```

```python
# [4] 데이터 확인
# - 기본 정보 확인
titanDF.info()
# age , embarked, deck,embark_town이 확인이 필요
```

```python
# 컬럼과 실제 데이터 타입 확인
print(titanDF.head(), titanDF.tail(), sep='\n\n')
```

```python
# 컬럼별 분포도 확인 #문자는 안나옴
titanDF.describe()
```

```python
# [5] 데이터 전처리
# - [5-1] 결측치 체크
titanDF.isna().sum() # age 177 , embarked 2, deck 2, embark_town 2
```

```python
# - [5-2] 중복데이터 체크
titanDF.duplicated().sum()
```

```python
titanDF[titanDF.duplicated()]
```

```python
du_titanDF=titanDF.drop_duplicates().copy()
```

```python
du_titanDF.info()
```

```python
du_titanDF['age']=du_titanDF['age'].fillna(du_titanDF['age'].mean())
du_titanDF['age']=np.floor(du_titanDF['age']).astype('int16') # numpy의 floor함수를 이용하여 반올림
```

```python
du_titanDF.info()
```

```python
du_titanDF.dropna(axis=0,subset=['embarked','embark_town'],inplace=True) #데이터삭제
```

```python
du_titanDF.info()
```

```python
# -[5-5] 자료형 변환 표준화
du_titanDF['survived']=du_titanDF['survived'].astype('category')
du_titanDF['pclass']=du_titanDF['pclass'].astype('category')
du_titanDF['sex']=du_titanDF['sex'].astype('category')
du_titanDF['class']=du_titanDF['class'].astype('category')
du_titanDF['adult_male']=du_titanDF['adult_male'].astype('category')
du_titanDF['embark_town']=du_titanDF['embark_town'].astype('category')
du_titanDF['alive']=du_titanDF['alive'].astype('category')


```

```python
custom_bins = [0, 18, 35, 50, 65, 100]
ages=pd.cut(du_titanDF['age'], bins=custom_bins, labels=['청소년','청년','중년','장년','노년'])
du_titanDF['ages']=ages
```

```python
du_titanDF
```

```python
du_titanDF['ages'].value_counts()
# 탑승자중 가장 많은 수가 청년이다.
```
