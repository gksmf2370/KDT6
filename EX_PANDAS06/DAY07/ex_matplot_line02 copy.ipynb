```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# [2] 데이터 준비
FILE_DATA = r'C:\Users\KDP15\Desktop\KDT6\EX_PANDAS06\DATA\남북한발전전력량.xlsx'
```

```python
# [3] Excel => DataFrame형태로 저장
dataDF=pd.read_excel(FILE_DATA)
```

```python
# [4] 데이터 확인
dataDF.info()
print(dataDF.head())
```

```python
# [5] 데이터 전처리 및 가공
# - 멀티인덱싱 : 컬럼 ==> 행인덱스 설정
# - (1) NaN컬럼에 대한 처리
# - (2) 두개의 컬럼을 행인덱스 설정
```

```python
# - (1) NaN컬럼에 대한 처리
dataDF[dataDF.columns[0]]
```

```python
# 전 데이터로 결측치를 채움
dataDF[dataDF.columns[0]].ffill(inplace=True)
```

```python
dataDF.head()
```

```python
# 행인덱스 사용할 컬럼명 추출
dataDF.columns[:2].to_list()
```

```python
# - (2) 두개의 컬럼을 행인덱스로 설정
dataDF.set_index([dataDF.columns[0],dataDF.columns[1]],inplace=True)
print(f'[확인]\n{dataDF.head(2)}')
```

```python
# - 행 인덱스 names 속성 수정
idx_names=dataDF.index.names
print(f'idx_names : {idx_names}')

dataDF.index.set_names({idx_names[0] : '구분', idx_names[1]:'전력'}, inplace=True)
print(f'[확인]\n{dataDF.head()}')
```

```python
# [5] 데이터 전처리 및 가공
# - '-' 데이터 값 NA로 치환 -->0으로 치환
dataDF.replace('-', pd.NA, inplace=True)
print(f'[확인]\n{dataDF.head()}')
```

```python
dataDF.isna().T.sum()
```

```python
dataDF.fillna(0, inplace=True)
dataDF.isna().T.sum()
```

```python
# - object ==> int16 형변환
dataDF=dataDF.astype('int16')
dataDF.dtypes
```

```python
# 남한과 북한의 합계 출력
plt.plot(dataDF.loc[('남한','합계')], 'bo--', label='S')
plt.plot(dataDF.loc[('북한','합계')], 'ro--', label='N')
plt.title('[남한 & 북한 발전량]')
plt.xlabel('YEAR')
plt.ylabel('DATA')
plt.legend(loc='upper left')
plt.show()
```

```python
def drawLineGraph(title, columns, labels, xlabel, ylabel):
    # 남한과 북한의 합계 출력
    # 라벨을 컬렴명으로 가져가는것
    for col in columns:
        plt.plot(dataDF.loc[col], label='col')
    
    # 이경우 라벨을 컬럼명 멀티인덱스로 가능
    #for idx, col in enumerate(columns):
    #    plt.plot(dataDF.loc[col], 'bo-', label=f'{labels[idx]}')
    plt.title(f'{title}')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(labels)
    plt.tight_layout()
    plt.show()
```

```python
drawLineGraph("TEST", [('남한','수력'), ('북한','수력'),('남한','합계'),('북한','합계')],['남한-수력','북한-수력','남한-합계','북한-합계'],'년도','발전량')
```

```python
# -> 하나의 figure에서 수력/화력/원자력/합계 그래프 넣기 1
fig=plt.figure() 

# - add_subplot(행수, 열수, 번호)
ax1=fig.add_subplot(2,2,1) # 2행 2열의 첫번째
ax2=fig.add_subplot(2,2,2) # 2행 2열의 두번째
ax3=fig.add_subplot(2,2,3) # 2행 2열의 세번째
ax4=fig.add_subplot(2,2,4) # 2행 2열의 네번째

# - 각 subplot에 값 넣기
ax1.plot(dataDF.loc[('남한','수력')], 'bo-', label='남한 - 수력')
ax2.plot(dataDF.loc[('남한','화력')], 'bo-', label='남한 - 화력')
ax3.plot(dataDF.loc[('남한','원자력')], 'bo-', label='남한 - 원자력')
ax4.plot(dataDF.loc[('남한','합계')], 'bo-', label='남한 - 합계')

# - 각 subplot별로 설정
ax1.set_title('남한 수력 발전량')
ax1.set_xlabel('년도')

ax2.set_title('남한 화력 발전량')
ax2.set_xlabel('년도')

ax3.set_title('남한 원자력 발전량')
ax3.set_xlabel('년도')

ax4.set_title('남한 합계 발전량')
ax4.set_xlabel('년도')

# - 화면에 띄우기
plt.tight_layout()
plt.legend()
plt.show()
```
