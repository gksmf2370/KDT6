## [ 남북한발전전력량 분석 및 시각화]
- 데이터셋 : 남북한발전전력량.xlsx
- 데이터구성
    * 1990~2016년까지 수력,화력,원자력,신재생 에너지 데이터
- 분석 주제 : 남한과 북한의 에너지 데이터 변화 분석 #비교하고싶은것

```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# [2] 데이터 준비
FILE_DATA = r'C:\Users\KDP15\Desktop\KDT6\EX_PANDAS06\DATA\남북한발전전력량.xlsx'
```

```python
# [3] Excel => DataFrame형태로 저장
dataDF=pd.read_excel(FILE_DATA)
```

```python
# [4] 데이터 확인
dataDF.info()
print(dataDF.head())
```

```python
# [5] 데이터 전처리 및 가공
# - 멀티인덱싱 : 컬럼 ==> 행인덱스 설정
# - (1) NaN컬럼에 대한 처리
# - (2) 두개의 컬럼을 행인덱스 설정
```

```python
# - (1) NaN컬럼에 대한 처리
dataDF[dataDF.columns[0]]
```

```python
# 전 데이터로 결측치를 채움
dataDF[dataDF.columns[0]].ffill(inplace=True)
```

```python
dataDF.head()
```

```python
# - (2) 두개의 컬럼을 행인덱스로 설정
dataDF.set_index([dataDF.columns[0],dataDF.columns[1]],inplace=True)
```

```python
dataDF
```

```python
dataDF.xs('신재생', level=1)
```

```python
dataDF.replace('-',0,inplace=True)
```

```python
dataDF.value_counts()
```

```python
dataDF.info()
```

```python
dataDF.loc[('북한','신재생'),:] = 0
```

```python
dataDF=dataDF.astype('int64')
```

```python
dataDF.info()
```

```python
dataDF.loc['남한']
```

```python
dataDF.loc['북한']
```

```python
# 남한과 북한의 합계 출력
plt.plot(dataDF.loc[('남한','합계')], 'bo--', label='S')
plt.plot(dataDF.loc[('북한','합계')], 'ro--', label='N')
plt.title('[SUM]')
plt.xlabel('YEAR')
plt.ylabel('DATA')
plt.legend(loc='upper left')
```

```python
# 남한과 북한의 신재생 출력
plt.plot(dataDF.loc[('남한','신재생')], 'bo--', label='S')
plt.plot(dataDF.loc[('북한','신재생')], 'ro--', label='N')
plt.title('[SUM]')
plt.xlabel('YEAR')
plt.ylabel('DATA')
plt.legend(loc='upper left')
```

```python
# 남한과 북한의 원자력 출력
plt.plot(dataDF.loc[('남한','원자력')], 'bo--', label='S')
plt.plot(dataDF.loc[('북한','원자력')], 'ro--', label='N')
plt.title('[SUM]')
plt.xlabel('YEAR')
plt.ylabel('DATA')
plt.legend(loc='upper left')
```
