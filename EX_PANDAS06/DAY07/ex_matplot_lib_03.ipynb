## [ Matplotlib 활용 데이터 시각화]
- x축과 Y축 다양한 처리

```python
# [1] 모듈 로딩
from matplotlib import pyplot as plt
```

```python
# [2] 그래프 기리기 -기본
y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']
plt.plot(x_data,y_data, marker='o')
plt.show()
```

```python
# [2] 그래프 기리기 - x축과 y축 범위설정
y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']
plt.plot(x_data,y_data, marker='o')
plt.ylim(bottom=0, top=15)
plt.xlim(left=0 , right=5)
plt.show()
```

```python
# [2] 그래프 기리기 - x축과 y축 눈금(tick) 설정
y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']
plt.plot(x_data,y_data, marker='o')
plt.xticks(rotation=45)
plt.show()
```

```python
# [2] 그래프 기리기 - 그래프 출력 창 즉, figure 크기 설정
y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

plt.figure(figsize=[10,10]) # 원하는 창의 크기 설정
plt.plot(x_data,y_data, marker='o')
plt.ylim(bottom=0, top=15)

plt.show()
```

```python
# [2] 그래프 기리기 - 그래프 출력 창 즉, figure 자동 맞춤
y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']


plt.plot(x_data,y_data, marker='o')
plt.ylim(bottom=0, top=15)
plt.tight_layout()   # 모든 내용 출력될수있도록 자동 맞춤 설정

plt.show()
```

```python
# [2] 그래프 기리기 - 그래프 출력 
y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']
plt.plot(x_data,y_data, marker='o')
plt.ylim(bottom=0, top=15)
plt.show()
```

```python
# [3] 그래프 기리기 - 눈금(ticks) 설정
y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']


plt.plot(y_data, marker='o')
plt.ylim(bottom=-1, top=10)
plt.xlim(left=-1, right=6)

font_conig={'fontsize':22, 'fontweight':'bold'}
plt.xticks([0,2,4,6], ['A','B','C','D'], **font_conig)
plt.show()
```
