## [ 파이 그래프 ]
- 컬럼(변수)의 데이터의 분포 비율을 시각화

```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# [2] 데이터 준비
F_DATA=r'C:\Users\KDP15\Desktop\KDT6\EX_PANDAS06\DATA\auto_mpg.csv'
```

```python
# [3] csv => DataFrame으로 저장
dataDF=pd.read_csv(F_DATA)
```

```python
# - 컬럼별 타입 확인
dataDF.dtypes
```

```python
# [4] 특정 컬럼의 데이터 비율
# - 자동차 생산국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
# - 자동차 생산국가에 대한 비율
round(dataDF['origin'].value_counts()/dataDF['origin'].count(),2)
#round(dataDF['origin'].value_counts(normalize=True),2)
```

```python
orginSR=dataDF['origin'].value_counts()
x=orginSR.index.to_list()
h=orginSR.values.tolist()
x, h
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - bar 그래프
orginBar=plt.bar(x, h, color=['red','yellow','green'], edgecolor=['blue']) # color은 그래프색깔넣기 edgecolor는 테두리색깔넣기

# 막대에 눈금설정
plt.xticks([1,2,3],['us','EU','JP'],rotation=45, fontsize=20)

# 막대에 라벨설정
plt.bar_label(orginBar, h, label_type='center')  # 막대위에 249, 70, 79표시
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['US','EU','JP'], autopct='%.1f%%')  #'.1f%%' 이건 소숫점 첫째자리를 의미함
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['US','EU','JP'], autopct='%.1f%%', 
        counterclock=False, startangle=90)  #'.1f%%' 이건 소숫점 첫째자리를 의미함
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['US','EU','JP'], autopct='%.1f%%', 
        counterclock=False, startangle=90, explode=[0, 0, 0.1])  #'.1f%%' 이건 소숫점 첫째자리를 의미함
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['US','EU','JP'], autopct='%.1f%%', 
        counterclock=False, startangle=90, wedgeprops={'linewidth':7, 'width':0.8,'edgecolor':'w'})
        #wedgeprops={'linewidth':2, 'width':0.5,'edgecolor':'r'}  이건 가운데 빈원과 관련된것임 , linewwidth는 굵기 , width는 가운데 원크기
plt.show()
```
