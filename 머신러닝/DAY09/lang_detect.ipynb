```python
import pandas as pd
import os
from collections import Counter
```

```python
# 데이터 파일
TRAIN_PATH = '언어판독데이터셋/train/'
TEST_PATH = '언어판독데이터셋/test/'

TRAIN_CSV = '../data/lang_train.csv'
TEST_CSV = '../data/lang_test.csv'
```

```python
# 파일 읽기 => open() -> read(), readline(), readlines() -> close()
#          => with open() as 구문
import os
if not os.path.exists(TRAIN_CSV):
    print(f'{TRAIN_CSV}파일이 없습니다.')
else:
    with open(TRAIN_CSV, mode='r') as f:
        alldata=f.read()
        print(f'[File Data] {alldata}')
```

```python
# 폴더 안에 파일 및 폴더 을 가져오기
files = os.listdir(TRAIN_PATH)

for file in files:
    print(file, TRAIN_PATH+file, os.path.exists(TRAIN_PATH+file))
    # 라벨 (en,fr,id,tl) 구하기
    label =file[:2]
    print(f'Label => {label}')

    with open(TRAIN_PATH+file, mode='r') as f:
        alldata=f.read() # 여기 첫번째 file의 내용(글)이 리스트로 다들어감 그래서 공백이 포함되어있다.
        #print(f'[File Data] {f.read()}')
        #print(f'alldata => {alldata}')

        # 대소문자 통일
        alldata = alldata.lower()
        print(f'alldata=> {len(alldata)}') 

        # 알파벳 아닌 문자들 제거
        alldata= alldata.replace('\n', '') # 줄바꿈을 제거
        print(f'alldata => {alldata}')
        print(f'공백제거 alldata=> {len(alldata)}') # 

        for ch in alldata:
            if ord('a') > ord(ch) or ord(ch) > ord('z') : #a~z 범위 밖인것들을 제거
                alldata=alldata.replace(ch, '')
        print(f'영어 alldata=> {len(alldata)}')

        with open(TRAIN_CSV, mode='a') as f:
            # - a ~ z 개수 카운팅
            cc=dict(Counter(alldata))
            f.write(label+',')
            f.write(','.join( [str(value) for value in cc.values() ]) + '\n')

```

```python
## 함수로 만들었음 a~z만 넣는함수
def make_dataset(dir_path, file_path):
    files = os.listdir(dir_path)

    for file in files:
        #print(file, dir_path+file, os.path.exists(dir_path+file))
        # 라벨 (en,fr,id,tl) 구하기
        label =file[:2]
        #print(f'Label => {label}')

        with open(dir_path+file, mode='r') as f:
            alldata=f.read() # 여기 첫번째 file의 내용(글)이 리스트로 다들어감 그래서 공백이 포함되어있다.
            #print(f'[File Data] {f.read()}')
            #print(f'alldata => {alldata}')

            # 대소문자 통일
            alldata = alldata.lower()
            #print(f'alldata=> {len(alldata)}') 

            # 알파벳 아닌 문자들 제거
            alldata= alldata.replace('\n', '') # 줄바꿈을 제거
            #print(f'alldata => {alldata}')
            #print(f'공백제거 alldata=> {len(alldata)}') # 

            for ch in alldata:
                if ord('a') > ord(ch) or ord(ch) > ord('z') : #a~z 범위 밖인것들을 제거
                    alldata=alldata.replace(ch, '')
            #print(f'영어 alldata=> {len(alldata)}')

            with open(file_path, mode='a') as f:
                # - a ~ z 개수 카운팅 ==> 정렬, 전체 단어수로 나누기
                #cc=dict(Counter(alldata))
                a_z=[]             
                for _ in range(ord('a'), ord('z')+1):    #여기서 이미 정렬을 해줌
                    cnt = alldata.count(chr(_)) 
                    a_z.append(str(cnt/len(alldata)))    #전체 단어수로 나눠줌

                f.write(label+',')
                f.write(','.join(a_z) + '\n')

```

```python
make_dataset(TRAIN_PATH, TRAIN_CSV)
make_dataset(TEST_PATH, TEST_CSV)
```
